```python
'''PROYECTO LÓGICA: Katas de Python
1. Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias 
de cada letra en la cadena. Los espacios no deben ser considerados.'''
```

```python
def frecuencias_letras(cadena):        #función llamada frecuencias_letras
    cadena = cadena.replace(" ", "")  # eliminar espacios
    frecuencias = {}                          # crear diccionario vacío para almacenar las frecuencias de las letras

    for letra in cadena:                    # Iterar sobre cada letra en la cadena
        if letra in frecuencias:            # Incrementar el contador de la letra en el diccionario
            frecuencias[letra] += 1
        else:
            frecuencias[letra] = 1

    return frecuencias                      # Devolver el diccionario con las frecuencias de las letras
print(frecuencias_letras("Hola mundo"))  # Ejemplo de uso de la función
```

```python
'''2.Dada una lista de números, obtén una nueva lista con el doble de cada valor. Usa la función 
map()'''
lista=[1, 2, 3, 4, 5]
doubled_list = list(map(lambda x: x * 2, lista))  # Aplicar la función lambda a cada elemento de la lista
print(doubled_list)  # Imprimir la lista resultante
```

```python
''' 3.Escribe una función que tome una lista de palabras y una palabra objetivo como parámetros. La función debe 
devolver una lista con todas las palabras de la lista original que contengan la palabra objetivo'''
```

```python
def buscar_palabras(lista_palabras, objetivo):
    resultado = []

    for palabra in lista_palabras:
        if objetivo in palabra:
            resultado.append(palabra)

    return resultado
print(buscar_palabras(["manzana", "banana", "naranja", "pera"], "pera"))  # Ejemplo de uso de la función
```

```python
''' 4.Genera una función que calcule la diferencia entre los valores de dos listas. Usa la función 
map()'''
```

```python
lista1 = [10, 20, 30]
lista2 = [1, 5, 7]
diferencia = list(map(lambda x, y: x - y, lista1, lista2))
print(diferencia)  # Imprimir la lista resultante
```

```python
'''5. Ecribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado, que por 
defecto es 5. La función debe calcular la media de los números en la lista y determinar si la media es mayor o igual 
que nota aprobado. Si es así, el estado será "aprobado", de lo contrario, será "suspenso". La función debe devolver 
una tupla que contenga la media y el estado.'''
```

```python
def evaluar_media(numeros, nota_aprobado=5):        #función llamada evaluar_media
    media = sum(numeros) / len(numeros)     # Calcular la media de los números en la lista

    if media >= nota_aprobado:      # Determinar si la media es mayor o igual que nota_aprobado
        estado = "aprobado"
    else:
        estado = "suspenso"

    return (media, estado)
print(evaluar_media([6, 7, 8, 9]))  # Ejemplo de uso de la función con la media mayor que nota_aprobado
```
